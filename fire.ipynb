```python
import pandas as pd
import numpy as np
import plotly.io as pio
pio.renderers

```

```python
df=pd.read_csv("./fire-occurence.csv")
print(df.shape)
df.head(8)
```

```python
df.columns
```

```python
df.loc[:,'FO_LandOwnType':'ModifiedDate'].head(5)
```

```python
print(df["CauseBy"].unique())
df["GeneralCause"].unique()
```

```python
df=df.loc[:,['Serial', 'DistrictName',"LandmarkLocation", "FireYear", 'Size_class', 'EstTotalAcres',
       'GeneralCause', 'Lat_DD', 'Long_DD', 'FO_LandOwnType', 'Ign_DateTime', 
       'Control_DateTime']]
df
```

```python
df.info()
```

```python
df=df.dropna()
df.info()
```

```python
df["Ign_DateTime"]=pd.to_datetime(df.loc[:,"Ign_DateTime"])
df["Control_DateTime"]=pd.to_datetime(df.loc[:,"Control_DateTime"])
df["Ign_DateTime"]=df["Ign_DateTime"].apply(lambda x:x.strftime('%Y-%m'))
df["Control_DateTime"]=df["Control_DateTime"].apply(lambda x:x.strftime('%Y-%m'))

df
```

```python
sizegrouper=df.groupby(["Size_class","FireYear"]).count()
sizegrouper=sizegrouper.drop(["DistrictName","LandmarkLocation","EstTotalAcres","GeneralCause","Lat_DD","Long_DD","FO_LandOwnType","Ign_DateTime","Control_DateTime"],axis=1)
sizegrouper
```

```python
sizegrouper1=sizegrouper.reset_index(level=[0,1])
sizegrouper1=sizegrouper1.set_index("FireYear")
sizegrouper1.head(5)
```

NUMBER OF FIRES DEPENDING ON THE SIZE

```python
import plotly.express as px

fig = px.line(sizegrouper1, x=sizegrouper1.index, y="Serial", color='Size_class')
fig.show(renderer="svg")
```

```python
df=df.sort_values(by="FireYear")
```

```python
dfdanger=df[(df["Size_class"]!="A")&(df["Size_class"]!="B")]
dfdanger=dfdanger[dfdanger["GeneralCause"]!="Lightning"]
```

```python
sizegrouper2=dfdanger.groupby(["Size_class","FireYear"]).count()
sizegrouper2=sizegrouper2.drop(list(sizegrouper2.columns[1:]),axis=1)
sizegrouper2=sizegrouper2.reset_index(level=[0,1])
sizegrouper2=sizegrouper2.set_index("FireYear")
sizegrouper2
```

NUMBER OF HUMAN DANGEROUS FIRES PER YEAR

```python
fig = px.line(sizegrouper2, x=sizegrouper2.index, y="Serial", color='Size_class')
fig.show(renderer="svg")
```

```python
acresy=df.groupby(["FireYear","DistrictName"]).sum().drop(["Serial","Lat_DD","Long_DD"], axis=1)
acresy=acresy.reset_index(level=[0,1])
acresy=acresy.set_index("FireYear")
acresy.head(7)
```

ACRES BURNED PER DISTRICT

```python
import plotly.graph_objects as go

fig = go.Figure(go.Bar(x=acresy.index.unique(), y=acresy[acresy["DistrictName"]==acresy["DistrictName"].unique()[0]]["EstTotalAcres"].values, name=acresy["DistrictName"].unique()[0]))

for a in range(len(acresy["DistrictName"].unique()[1:])):

    fig.add_trace(go.Bar(x=acresy.index.unique(), y=acresy[acresy["DistrictName"]==acresy["DistrictName"].unique()[a]]["EstTotalAcres"].values, name=acresy["DistrictName"].unique()[a]))

fig.update_layout(barmode='stack')
fig.update_xaxes(categoryorder='category ascending')
fig.show(renderer="svg")
```

```python
landtype=df.groupby(["FO_LandOwnType","FireYear"]).count()
landtype=landtype.drop(list(landtype.columns[1:]),axis=1)
landtype=landtype.reset_index(level=[0,1])
landtype=landtype.set_index("FireYear").sort_index()
landtype.rename(columns={"Serial":"Number_Fires"}, inplace=True)
landtype
```

```python
landtype["FO_LandOwnType"].unique()
```

```python
import plotly.graph_objects as go

fig = go.Figure(go.Bar(x=landtype.index.unique(), y=landtype[landtype["FO_LandOwnType"]==landtype["FO_LandOwnType"].unique()[0]]["Number_Fires"].values, name=landtype["FO_LandOwnType"].unique()[0]))

for a in range(len(landtype["FO_LandOwnType"].unique()[1:])):

    fig.add_trace(go.Bar(x=landtype.index.unique(), y=landtype[landtype["FO_LandOwnType"]==landtype["FO_LandOwnType"].unique()[a]]["Number_Fires"].values, name=landtype["FO_LandOwnType"].unique()[a]))

fig.update_layout(barmode='stack')
fig.update_xaxes(categoryorder='category ascending')
fig.show(renderer="svg")
```

```python
fig = px.line(landtype, x=landtype.index, y="Number_Fires", color='FO_LandOwnType')
fig.show(renderer="svg")
```
